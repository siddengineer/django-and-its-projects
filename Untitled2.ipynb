```python
df['Total_Jan_to_May'] = df[['Jan','Feb','Mar','April','May']].sum(axis = 1)
print(df[['Year','Total_Jan_to_May']].head())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(df['Year'],df['Total_Jan_to_May'],marker ='o',color = 'blue')
plt.title('Total Rainfall from Jan to May(mumbai)')
plt.xlabel('Year')
plt.ylabel('Rainfall(mm)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('mumbai-monthly-rains.csv')
print(df.columns.tolist())
df['Monsoon'] = df[['June','July','Aug','Sept']].sum(axis = 1)
df['Monsoon_10yr_avg'] = df['Monsoon'].rolling(10).mean()

plt.figure(figsize=(10,5))
plt.plot(df['Year'],df['Monsoon'],alpha = 0.3,label = 'Actual')
plt.plot(df['Year'],df['Monsoon_10yr_avg'],color = 'blue',label = '10-Year Rolling Avg')
plt.title("Monsoon Rainfall Trends(1901-2020)")
plt.xlabel('Year')
plt.ylabel('Rainfall(mm)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df['Decade'] = (df['Year'] // 10)*10
decade_avg = df.groupby('Decade')['Monsoon'].mean().reset_index()
plt.figure(figsize=(10,5))
plt.bar(decade_avg['Decade'],decade_avg['Monsoon'], width = 8,color ='skyblue')
plt.title("Average Monsoon Rainfall Decade-wise")
plt.xlabel("Decade")
plt.ylabel("Average Rainfall(mm)")
plt.grid(True)
plt.show
```
